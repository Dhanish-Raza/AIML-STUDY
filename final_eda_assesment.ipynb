Compare renewable vs non-renewable plants
(Solar, Wind, Hydro vs Coal, Gas).


```python
df.head()
```

```python
renewable=["Solar","Wind","Hydro"]
df["Energy_Type"]=df["Fuel_Type"].apply(lambda x: "Renewable" if x in renewable else "Non-Renewable")
```

```python
df.head(5)
```

Create a summary table with:
â€¢	Fuel_Type
â€¢	Avg Capacity
â€¢	Avg Efficiency
â€¢	Avg COâ‚‚ Emissions


```python
table = df.groupby("Fuel_Type").agg(
    Avg_Capacity=("Capacity_MW", "mean"),
    Avg_Efficiency=("Efficiency_%", "mean"),
    Avg_CO2_Emissions=("CO2_Emissions_tons", "mean"))
print(table)   
```
