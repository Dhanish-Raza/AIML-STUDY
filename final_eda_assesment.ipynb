Find number of power plants per fuel type.

```python
df
```

```python
no_power_plant = df['Fuel_Type'].value_counts()
print(no_power_plant)
```

Find average capacity (MW) by fuel type.

```python
df
```

```python
avg_capacity = df.groupby("Fuel_Type")["Capacity_MW"].mean()
print(avg_capacity)
```

Which country has the highest total energy generation?

```python
total_generation = df.groupby("Country")["Annual_Generation_GWh"].sum()
print(total_generation)
```

Compare average efficiency across regions.

```python
df
```

```python
avg_efficiency_accros_region = df.groupby("Region")["Efficiency_%"].mean()
print(avg_efficiency_accros_region)
```

Count how many plants are Active vs Maintenance.

```python
df
```

```python
count_active_maintenance = df["Plant_Status"].value_counts()
print(count_active_maintenance)
```

Find average CO₂ emissions per fuel type.

```python
df
```

```python
avg_co2 = df.groupby("Fuel_Type")["CO2_Emissions_tons"].mean()
print(avg_co2)
```

Create a new column Plant_Age

```python
df
```

```python
df["Plant_Age"]=df["Commission_Year"].apply(lambda x: 2026-x)

```

```python
df
```

Find average plant age by fuel type.

```python
avg_plant_age = df.groupby("Fuel_Type")["Plant_Age"].mean()
print(avg_plant_age)
```

Which fuel type produces the most energy per MW?

```python
df.head()
```

```python
df["Energy_per_MW"] = df["Annual_Generation_GWh"] / df["Capacity_MW"]

energy_per_mw = df.groupby("Fuel_Type")["Energy_per_MW"].mean()
print(energy_per_mw.idxmax(), energy_per_mw.max())
```

```python
most_eng_MW = df.groupby("Fuel_Type")['Capacity_MW'].max()
print(most_eng_MW)
```

Compare renewable vs non-renewable plants
(Solar, Wind, Hydro vs Coal, Gas).


```python
df.head()
```

```python
df['Energy_Type'] = df['Fuel_Type'].apply(lambda x: 'Renewable' if x in ['Hydro', 'Wind', 'Solar'] else 'Non-Renewable')
df
```

Create a summary table with:
•	Fuel_Type
•	Avg Capacity
•	Avg Efficiency
•	Avg CO₂ Emissions


```python
table = df.groupby("Fuel_Type").agg(Avg_Capacity=("Capacity_MW", "mean"),Avg_Efficiency=("Efficiency_%", "mean"),Avg_CO2_Emissions=("CO2_Emissions_tons", "mean"))
print(table)   
```
