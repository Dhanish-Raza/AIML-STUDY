```python
# fill age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
# fill city with mode
df['City'].fillna(df['City'].mode()[0], inplace=True)
# fill Session_Duration with mean
df['Session_Duration'].fillna(df['Session_Duration'].mean(), inplace=True)
# fill Data_Used_MB with mean
df['Data_Used_MB'].fillna(df['Data_Used_MB'].mean(), inplace=True)
# fill 	App_Rating with mean
df['App_Rating'].fillna(df['App_Rating'].mean(), inplace=True)
```
