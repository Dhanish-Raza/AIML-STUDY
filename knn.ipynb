```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=np.array([
    [18, 20000],
    [20, 30000],
    [25, 50000],
    [30, 60000],
    [35, 80000],
    [40, 90000],
    [45, 110000],
    [50, 120000],
    [55, 140000],
    [60, 150000]
])
```

Scaling

```python
def min_max_scale(X):
    X_scaled = np.zeros_like(X, dtype=float)
    for col in range(X.shape[1]):
        min_val = np.min(X[:, col])
        max_val = np.max(X[:, col])
        X_scaled[:, col] = (X[:, col] - min_val) / (max_val - min_val)
    return X_scaled
X_scaled = min_max_scale(X)
print("Scaled Data:\n", X_scaled)
```

```python
def min_max_scaling(X):
    X_min = X.min(axis=0)
    X_max = X.max(axis=0)
    X_scaled = (X - X_min) / (X_max - X_min)
    return X_scaled
```
