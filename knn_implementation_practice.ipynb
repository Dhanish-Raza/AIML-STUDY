```python
import numpy as np
```

```python
X=np.array([
    [18,20000],
    [25,30000],
    [35,35000],
    [40,80000],
    [60,100000]
])
```

```python
def min_max_scale(X):
  X_scaled=np.zeros_like(X,dtype=float)

  for col in range(X.shape[1]):
    min_val=X[:,col].min()
    max_val=X[:,col].max()

    X_scaled[:,col]=(X[:,col]-min_val)/(max_val-min_val)

  return X_scaled
X_min_max=min_max_scale(X)
print(X_min_max)
```

```python
def euclidean_distance(p1,p2):
  return np.sqrt(np.sum(p1-p2)**2)
```

```python
p1=X[0]
p2=X[1]
print(euclidean_distance(p1,p2))
```

```python
p1_scaled=X_min_max[0]
p2_scaled=X_min_max[1]
print(euclidean_distance(p1_scaled,p2_scaled))
```

```python
def standard_scaler(X):
    X_scaled = np.zeros_like(X, dtype=float)

    for col in range(X.shape[1]):
        mean = X[:, col].mean()
        std = X[:, col].std()

        X_scaled[:, col] = (X[:, col] - mean) / std

    return X_scaled

x_standard=standard_scaler(X)
print(x_standard)
```

```python
X=np.array([
    [18,20000],
    [25,30000],
    [35,35000],
    [40,80000],
    [60,100000]
])

Y=np.array([0,0,0,1,1])
```

```python
def knn_predict(X_train,y_train,x_test,k):
  distances=[]

  for i in range(len(X_train)):
    dist=euclidean_distance(X_train[i],x_test)
    distances.append((dist,y_train[i]))

  distances.sort(key=lambda x:x[0])
  neighbors=distances[:k]

  labels=[label for _,label in neighbors]

  return max(set(labels),key=labels.count)
```

```python
x_test=np.array([45,90000])

pred=knn_predict(X,Y,x_test,k=3)
print(pred)
```
